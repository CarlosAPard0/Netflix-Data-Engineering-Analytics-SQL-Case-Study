```python
## Cargar datos
import pandas as pd
df = pd.read_csv('Data/netflix_titles.csv')
```

```python
## Conectar a servidor local sql ssms
import sqlalchemy as sal
from sqlalchemy.types import NVARCHAR 
from sqlalchemy.dialects.mssql import VARCHAR
cadena_de_coneccion = (
    "mssql://localhost\\SQLEXPRESS/master" 
    "?driver=ODBC+Driver+18+for+SQL+Server"
    "&TrustServerCertificate=yes"
)
engine = sal.create_engine(cadena_de_coneccion)
conn = engine.connect()
```

```python
## Cargar dataset a ssms y cambiar tipos de datos
tipos_sql = {
    'title': NVARCHAR(length=200),
    'show_id': VARCHAR(length=10),
    # Puedes omitir el resto si no tienen caracteres especiales, pero es buena práctica
}
df.to_sql('netflix_raw', con=conn, index = False, if_exists = 'replace', dtype=tipos_sql)
conn.close()
```
